```python
import matplotlib.pyplot as plt

candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'THáx']
candy_counts = [52, 39, 78, 13, 78]

# Membuat diagram pie
plt.pie(candy_counts, labels=candy_names, autopct='%1.1f%%')
plt.title('Persentase pengambilan permen')
plt.show()

total_candies = sum(candy_counts) 
snickers_percentage = (candy_counts[candy_names.index('Snickers')] / total_candies) *100
print("Persentase Snickers diambil :", snickers_percentage, "%")
```
